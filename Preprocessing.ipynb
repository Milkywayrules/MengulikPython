```python
import nltk
import string
import re
# import StemmerFactory class
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
# import StopWordRemoverFactory class
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

import matplotlib.pyplot as plt

import pandas as pd
```

```python
# [^\P{P}\-.]+


def case_folding(text):
    return text.lower()

def strip_number(text):
    return re.sub(r"\d+", "", text)

# def strip_punctuation(text):
#     return text.translate(str.maketrans("","",string.punctuation))

def strip_whitespace(text):
    return re.sub('\s+',' ',text.strip())

def remove_punc_num(text):
    text = re.sub(r'([.*]){2,}','.',text)
    return re.sub(r'([^a-z\s])(([\-,.*]){2,})','',text)
#     return re.sub(r'[^a-z\s.]',' ',text)

def sentence_tokenizing(text):
    return nltk.tokenize.sent_tokenize(text)

def word_tokenizing(text):
    return nltk.tokenize.word_tokenize(text)
```

```python
f = open("txt/dilan.txt", "r")
text = f.read()

text = 'Dari kantin, sebelum mau masuk ke kelas, aku berpapasan dengan dilan, dia jalan bersama kawan-kawannya... Pasti baru datang dari warung bi eem, "milea!" dia memanggil dan mendekat, kuhentikan langkahku. Nandan, Hadi, dan Rani terus berjalan karena aku minta untuk duluan. "Boleh gak aku ikut pelajaran di kelasmu lagi?" dia senyum, aku juga. "Kamu mau bikin aku senang gak?" ku pandang matanya hampir-hampir gak percaya bahwa aku bisa nanya seberani itu kepadanya!'

text = case_folding(text)
text = remove_punc_num(text)
text = strip_number(text)
text = strip_whitespace(text)

print(text)
```

```python
# menampilkan hasil tokenisasi kalimat
sentence_tokens = sentence_tokenizing(text)
print('Sentence tokenizer Result : \n') 
print(sentence_tokens)
print('\n\n')

# menampilkan hasil tokenisasi kalimat per baris
print('Sentence tokenizer Result (per row) : \n')
i = 1
for x in sentence_tokens:
    print(str(i) + '.) ' + x)
    i += 1
```

```python
# menampilkan hasil tokenisasi kata
word_tokens = word_tokenizing(text)
print('Word tokenizer Result : \n') 
print(word_tokens)
print('\n\n')


# menampilkan hasil tokenisasi kata per baris
# print('Word tokenizer Result (per row) : \n')
# i = 1
# for x in word_tokens:
#     print(str(i) + '.) ' + x)
#     i += 1
```

```python
def tokens_freq(tokens):
    return nltk.FreqDist(tokens)


# menampilkan hasil tokenisasi dengan frekuensi kemunculan
tokens_freq = tokens_freq(word_tokens)
print('Tokens frequency : \n') 
print(tokens_freq.most_common())
print('\n\n')

# # menampilkan hasil tokenisasi dengan frekuensi kemunculan per baris
# print('Tokens Frequency Result (per row) : \n')
# for x in tokens_freq.most_common():
#     print(x)
```
